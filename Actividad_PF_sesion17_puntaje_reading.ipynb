#2.	Crea / Presenta la forma que creas más conveniente para  describir cuántos alumnos por escuela tuvieron el peor puntaje  en "reading", y con ello saber que genero es el que predomina  en este ﬁltro.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
data_students=pd.read_csv("Clean_students_complete.csv")
```

```python
def Func1(x,y):
    minimo=data_students["reading_score"].min()
    maximo=data_students["reading_score"].max()
    col=list(data_students.columns)
    if y not in col:
        print("Esta asignatura no existe")
    elif x<minimo or x>maximo:
        print("Ningun estudiante obtuvo esa calificación")
    else:
        data_score_fun=data_students[data_students[y]==x]
        dummies_genero_fun=pd.get_dummies(data_score_fun["gender"])
        gender_scname_fun=data_score_fun[["school_name","gender"]]
        plotbygen_fun=pd.concat([gender_scname_fun,dummies_genero_fun],axis=1)
        agrupado_fun=plotbygen_fun.groupby(["school_name"]).sum()
        agrupado_fun.plot.bar(figsize=(10,10))
    
```

#### Crea / Presenta la forma que creas más conveniente para  describir cuántos alumnos por escuela tuvieron el mejor  puntaje en "reading", y con ello saber que genero es el que  predomina en este ﬁltro. 

```python
Func1(data_students["reading_score"].max(),"reading_score")
```

#### Crea / Presenta la forma que creas más conveniente para  describir cuántos alumnos por escuela tuvieron el peor puntaje  en "reading", y con ello saber que genero es el que predomina  en este ﬁltro.

```python
Func1(data_students["reading_score"].min(),"reading_score")
```
